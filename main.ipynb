```python
# Import required libraries

import chardet
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score
from sklearn.linear_model import Ridge, Lasso, ElasticNet, LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_squared_log_error
from sklearn.inspection import permutation_importance
from sklearn.linear_model import ElasticNet, Ridge, Lasso, LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from scipy.stats import uniform, loguniform

# Function to detect file encoding
def detect_file_encoding(file_path):
    with open(file_path, 'rb') as file:
        result = chardet.detect(file.read())
    return result['encoding']

# Function to remove gender related columns from dataframe
def remove_gender(df):
    columns_to_drop = [column for column in df.columns if 'M' in column or 'F' in column]
    return df.drop(columns_to_drop, axis=1)

# Function to remove correlated columns from dataframe
# Define a function named 'remove_correlated_columns' that takes a pandas DataFrame 'df' and a threshold value as inputs
def remove_correlated_columns(df, threshold=0.1):
    
    # Compute the correlation matrix of the input DataFrame 'df' using the corr() method, take the absolute value to get the magnitude of the correlation
    corr_matrix = df.corr().abs()
    
    # Generate a list of tuples containing pairs of columns with a correlation coefficient greater than the specified threshold
    high_corr_pairs = [(col1, col2) for col1 in corr_matrix.columns for col2 in corr_matrix.columns if corr_matrix.loc[col1, col2] > threshold and col1 != col2]
    
    # Extract the second column from each pair of 'high_corr_pairs', convert them to a set to remove duplicates, then convert it back to a list
    columns_to_drop = list(set([pair[1] for pair in high_corr_pairs]))
    
    # Drop the highly correlated columns from the original DataFrame 'df' by calling the drop() method and passing in the 'columns_to_drop' list and 'axis=1' to indicate that the columns are to be dropped
    return df.drop(columns_to_drop, axis=1)

#  Function to prepare dataframe with relevant columns
def prepare_data(df):
    
    # Create a new column in the DataFrame 'df' named 'Electoral Division', which is a concatenation of the last three characters of the 'GEOGID' column and the 'GEOGDESC' column
    df['Electoral Division'] = df['GEOGID'].str[-3:] + ' ' + df['GEOGDESC']
    
    # Rename the column 'T10_3_ENGT' to 'Highest Education Level' and the column 'T11_1_TW' to 'Total Hours Worked'
    df = df.rename(columns={'T10_3_ENGT': 'Highest Education Level', 'T11_1_TW': 'Total Hours Worked'})
    
    # Return a new DataFrame with only the columns 'Electoral Division', 'Highest Education Level', and 'Total Hours Worked'
    return df[['Electoral Division', 'Highest Education Level', 'Total Hours Worked']]

# Function to merge multiple dataframes on a common column
def merge_data(dfs, on='Electoral Division'):
    
    # Initialize a new DataFrame 'merged_df' as the first DataFrame in the list 'dfs'
    merged_df = dfs[0]
    
    # Iterate through the remaining DataFrames in the list 'dfs' and merge them with 'merged_df' on the specified column 'on'
    for df in dfs[1:]:
        merged_df = merged_df.merge(df, on=on)
    
    # Return the merged DataFrame
    return merged_df

# Function to keep the first three characters of a string
def keep_first_three_chars(column):
    
    return column[:3]

# Function to convert dataframe columns to int
def convert_to_int(df):
    
    # Iterate through each column in the DataFrame 'df'
    for column in df.columns:
        
        # Check if the column is numeric and not a float
        if pd.api.types.is_numeric_dtype(df[column]) and not pd.api.types.is_float_dtype(df[column]):
            
            # Convert the column to numeric data type, set missing values to the median of the column, and convert it to integer data type
            df[column] = pd.to_numeric(df[column], errors='coerce')
            median = df[column].median()
            df[column] = df[column].fillna(median)
            df[column] = df[column].astype(int)
        
        # If the column is not numeric, replace missing values represented as ".." with NaN and set missing values to the median of the column
        else:
            df[column] = df[column].replace("..", np.nan)
            median = df[column].median()
            df[column] = df[column].fillna(median)
    
    # Return the modified DataFrame
    return df


# Function to process the main CSV file
def process_main_csv():
    
    # Read the CSV file 'main.csv' into a pandas DataFrame 'df', specifying that the thousands separator is a comma
    df = pd.read_csv("main.csv", thousands=',')
    
    # Apply the 'keep_first_three_chars' function to the 'Electoral Division' column of the DataFrame 'df'
    df['Electoral Division'] = df['Electoral Division'].apply(keep_first_three_chars)
    
    # Modify the DataFrame 'df' by converting numeric columns to integer data type and handling missing values
    modified_df = convert_to_int(df)
    
    # Write the modified DataFrame 'modified_df' to a new CSV file named 'main_modified.csv', excluding the index column
    modified_df.to_csv('main_modified.csv', index=False)

# Function to remove outliers from dataframe using IQR method
def remove_outliers_iqr(df, target_col):
    
    # Compute the first quartile, third quartile, and interquartile range (IQR) of the target column using the quantile() method
    Q1 = df[target_col].quantile(0.25)
    Q3 = df[target_col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Compute the lower and upper bounds for outliers using the IQR
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filter the DataFrame 'df' to exclude rows where the target column is outside the lower and upper bounds
    df_out = df[(df[target_col] > lower_bound) & (df[target_col] < upper_bound)]
    
    # Return the filtered DataFrame
    return df_out


# Function to scale the features using StandardScaler
def scale_features(X_train, X_test):
    
    # Initialize a StandardScaler object 'scaler'
    scaler = StandardScaler()
    
    # Fit and transform the training data 'X_train' using the StandardScaler object 'scaler', and store the result in 'X_train_scaled'
    X_train_scaled = scaler.fit_transform(X_train)
    
    # Transform the testing data 'X_test' using the StandardScaler object 'scaler', and store the result in 'X_test_scaled'
    X_test_scaled = scaler.transform(X_test)
    
    # Return the scaled training and testing data as a tuple
    return X_train_scaled, X_test_scaled
    
# Function to split the data into train and test sets
def split_data(df, target_col, test_size=0.2, random_state=42):
    
    # Create a new DataFrame 'X' by dropping the target column from the input DataFrame 'df'
    X = df.drop(target_col, axis=1)
    
    # Create a new Series 'y' by selecting the target column from the input DataFrame 'df'
    y = df[target_col]
    
    # Split the data into training and testing sets using the train_test_split function, and store the results in four variables: X_train, X_test, y_train, y_test
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    
    # Return the training and testing sets as a tuple
    return X_train, X_test, y_train, y_test

# Function to train and evaluate a model
# This function also includes hyperparameter tuning for Ridge, Lasso, and ElasticNet models
def train_and_evaluate_model(model, X_train, X_test, y_train, y_test, model_name):
    
    # If the model name is Ridge, Lasso, or ElasticNet, create a dictionary 'param_distributions' with the hyperparameters to be tuned
    if model_name in ['Ridge', 'Lasso', 'ElasticNet']:
        param_distributions = {
            'alpha': uniform(loc=0.001, scale=10),
        }

        if model_name == 'ElasticNet':
            param_distributions.update({
                'l1_ratio': uniform(loc=0, scale=1)
            })

        # Create a RandomizedSearchCV object 'model' with the hyperparameters to be tuned
        model = RandomizedSearchCV(
            model,
            param_distributions,
            n_iter=100,
            cv=5,
            n_jobs=-1,
            verbose=0,
            random_state=42
        )

    # Train the machine learning model on the training data 'X_train' and 'y_train'
    model.fit(X_train, y_train)
    
    # Determine the best model based on the hyperparameters tuned in RandomizedSearchCV (if applicable)
    best_model = model.best_estimator_ if isinstance(model, RandomizedSearchCV) else model
    
    # Generate predictions on the testing data 'X_test'
    y_pred = best_model.predict(X_test)

    # Perform cross-validation on the best model
    cv_scores = cross_val_score(best_model, X_train, y_train, cv=5)

    # Compute the evaluation metrics (R^2, MSE, RMSE, MSLE, MAE) using the predictions and true values of the testing data 'y_test' and 'y_pred'
    r2 = r2_score(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    msle = mean_squared_log_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)

    # Return the evaluation metrics, predictions, best model, and cross-validation scores as a tuple
    return r2, mse, rmse, msle, mae, y_pred, best_model, cv_scores

# Function to plot actual vs predicted values
def plot_actual_vs_predicted(y_test, y_pred, model_name):
    
    # Create a scatter plot with blue markers for predicted values and red markers for true values
    plt.scatter(y_test, y_pred, color='b', label='Predicted', alpha=0.5)
    plt.scatter(y_test, y_test, color='r', label='Actual', alpha=0.5)
    
    # Set the labels and title of the plot
    plt.xlabel('Actual')
    plt.ylabel('Predicted')
    plt.title(f'{model_name}: Actual vs Predicted')
    
    # Show a legend for the plot
    plt.legend()
    
    # Display the plot
    plt.show()

# Function to analyze and visualize permutation importance
def analyze_and_visualize_permutation_importance(model, X_test, y_test, feature_names, model_name):
    
    # Compute permutation importances for the machine learning model 'model' using the testing data 'X_test' and 'y_test', and store the results in 'result'
    result = permutation_importance(model, X_test, y_test, n_repeats=10, random_state=42)
    
    # Create a DataFrame 'importance_df' to store the feature names, mean importances, and standard deviations of the permutation importances
    importance_df = pd.DataFrame({'Feature': feature_names, 'Importance Mean': result.importances_mean, 'Importance Std': result.importances_std})
    
    # Sort the DataFrame 'importance_df' by mean importance in descending order
    importance_df = importance_df.sort_values(by='Importance Mean', ascending=False)
    
    # Print the permutation importances for each feature
    print(f"{model_name} Permutation Importances:")
    print(importance_df)
    print()
    
    # Visualize permutation importances using a bar plot
    plt.figure(figsize=(10, 5))
    sns.barplot(data=importance_df, x='Feature', y='Importance Mean', yerr=importance_df['Importance Std'])
    plt.xticks(rotation=90)
    plt.title(f'{model_name}: Feature Permutation Importances')
    plt.show()


def analyze_and_visualize_coefficients(model, feature_names, model_name):
    
    # Retrieve the coefficients of the machine learning model 'model' and store them in 'coef'
    coef = model.coef_
    
    # Create a DataFrame 'coef_df' to store the feature names and coefficients
    coef_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': coef})
    
    # Sort the DataFrame 'coef_df' by the absolute values of the coefficients in descending order
    coef_df = coef_df.reindex(coef_df.Coefficient.abs().sort_values(ascending=False).index)
    
    # Print the coefficients for each feature
    print(f"{model_name} Coefficients:")
    print(coef_df)
    
    # Visualize coefficients using a bar plot
    plt.figure(figsize=(10, 5))
    sns.barplot(data=coef_df, x='Feature', y='Coefficient')
    plt.xticks(rotation=90)
    plt.title(f'{model_name}: Feature Coefficients')
    plt.show()
    
def perform_clustering(X, selected_features, n_clusters=4, random_state=42):
    
    # Create a KMeans clustering model with 'n_clusters' clusters and 'random_state' for reproducibility
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
    
    # Fit the KMeans model to the selected features of the feature matrix 'X' and predict the clusters
    clusters = kmeans.fit_predict(X[selected_features])
    
    # Return the predicted clusters
    return clusters
    
def visualize_clusters(df, cluster_col, target_col):
    
    # Visualize clusters vs target column using a box plot
    plt.figure(figsize=(10, 5))
    sns.boxplot(data=df, x=cluster_col, y=target_col)
    plt.title('Clusters vs Household Income')
    plt.xlabel('Cluster')
    plt.ylabel('Household Income')
    plt.show()

def main():
    # Read SAPS2016_ED3409.csv file and remove gender column
    df = pd.read_csv("SAPS2016_ED3409.csv", encoding='ISO-8859-1')
    df = remove_gender(df)
    df.to_csv("SAPS2016_ED3409_no_gender.csv", index=False)

    # Read the modified CSV file and remove correlated columns
    df = pd.read_csv("SAPS2016_ED3409_no_gender.csv")
    df = remove_correlated_columns(df)
    df.to_csv("SAPS2016_ED3409_reduced_columns.csv", index=False)

    # Read the reduced columns CSV file and prepare data
    df = pd.read_csv("SAPS2016_ED3409_reduced_columns.csv")
    df = prepare_data(df)
    df.to_csv('SAPS2016_ED3409_ready_to_merge.csv', index=False)

    # Merge different datasets
    ed_proportion_sector = pd.read_csv("EDProportionSector.csv", encoding='ISO-8859-1')
    ed_social_welfare = pd.read_csv("EDSocialWelfare.csv", encoding='ISO-8859-1')
    ed_pension = pd.read_csv("EDPension.csv", encoding='ISO-8859-1')
    ed_median_gross_income_sd_commuter = pd.read_csv("EDMedianGrossIncomeSDCommuter.csv", encoding='ISO-8859-1')
    ed_median_gross_income_ld_commuter = pd.read_csv("EDMedianGrossIncomeLDCommuter.csv", encoding='ISO-8859-1')
    ed_gross_income_2016 = pd.read_csv("EDGrossIncome2016.csv", encoding='ISO-8859-1')

    # Store the different dataframes in a list
    dfs = [ed_proportion_sector, ed_social_welfare, ed_pension, ed_median_gross_income_sd_commuter, ed_median_gross_income_ld_commuter, ed_gross_income_2016]
    # Merge all the dataframes
    merged_df = merge_data(dfs)
    merged_df.to_csv("MergedData.csv", index=False)

    # Read ready-to-merge dataset and merge with merged dataframe
    df1 = pd.read_csv("SAPS2016_ED3409_ready_to_merge.csv")
    df2 = pd.read_csv("MergedData.csv")

    # Extract only the electoral division from the electoral division column
    df2['Electoral Division Stripped'] = df2['Electoral Division'].apply(lambda x: x.split(',')[0].strip())
    # Merge two dataframes based on electoral division
    merged_df = pd.merge(df1, df2, left_on='Electoral Division', right_on='Electoral Division Stripped', suffixes=('_x', '_y'))

    # Drop unnecessary columns and rename columns
    merged_df.drop(['Electoral Division_x', 'Electoral Division Stripped'], axis=1, inplace=True)
    merged_df.rename(columns={'Electoral Division_y': 'Electoral Division'}, inplace=True)

    # Store the merged dataframe in a csv file
    merged_df.to_csv("main.csv", index=False)

    # Process the main csv file
    process_main_csv()
    
    # Load the modified dataset
    df = pd.read_csv('main_modified.csv')

    # Set the target column as 'Household Income'
    target_col = 'Household Income'

    # Remove outliers using the IQR method
    df = remove_outliers_iqr(df, target_col)

    # Split the dataset into 3 subsets, scale features, and store subsets in a list
    subsets = []
    for _ in range(3):
        X_train, X_test, y_train, y_test = split_data(df, target_col)
        # Store the column names before scaling
        feature_names = X_train.columns
        X_train, X_test = scale_features(X_train, X_test)
        subsets.append((X_train, X_test, y_train, y_test))

    # Choose relevant features for clustering
    selected_features = [
        '% Receiving SW',
        '% Pension is Majority',
        'Income per commuter for <30mins',
        'Income per commuter for >30mins'
    ]

    # Perform K-means clustering on the dataset with selected features
    n_clusters = 3  # Change this value to the number of clusters you want to use
    df['Cluster'] = perform_clustering(df, selected_features, n_clusters=n_clusters)

    # Analyze the relationship between the clusters and household income
    cluster_analysis = df.groupby('Cluster').agg(
        {
            target_col: ['mean', 'median', 'std'],
            'Electoral Division': 'count',
        }
    )
    print("Cluster Analysis:")
    print(cluster_analysis)

    # Visualize the relationship between the clusters and household income
    visualize_clusters(df, 'Cluster', target_col)

    # Train and evaluate 3 different regression models
    models = [LinearRegression(), Ridge(alpha=1.0), Lasso(alpha=1.0), ElasticNet(alpha=1.0, l1_ratio=0.5)]
    model_names = ['Linear Regression', 'Ridge', 'Lasso', 'ElasticNet']
    metrics = []
    best_models = []  # Store the best models for further analysis
    for model, model_name in zip(models, model_names):
        # Train and evaluate model
        r2, mse, rmse, msle, mae, y_pred, best_model, cv_scores = train_and_evaluate_model(model, X_train, X_test, y_train, y_test, model_name)
        best_models.append(best_model)
        metrics.append((r2, mse, rmse, msle, mae))
        # Print evaluation metrics
        print(f'{model_name} Metrics:')
        print(f'  R^2: {r2}\n  MSE: {mse}\n  RMSE: {rmse}\n  MSLE: {msle}\n  MAE: {mae}')
        print(f'  Cross-Validation Mean: {cv_scores.mean()}, Cross-Validation Std: {cv_scores.std()}')

        # Print the first 3 actual and predicted values
        print(f'\n{model_name} First 3 Actual vs Predicted Values:')
        for j in range(3):
            print(f'  Actual: {y_test.iloc[j]}, Predicted: {y_pred[j]}')

        # Plot actual vs predicted values
        plot_actual_vs_predicted(y_test, y_pred, model_name)

    # Analyze and visualize coefficients for best models
    for best_model, model_name in zip(best_models, model_names):
        if hasattr(best_model, 'coef_'):
            analyze_and_visualize_coefficients(best_model, feature_names, model_name)

    # Analyze and visualize permutation importance for best models
    for best_model, model_name in zip(best_models, model_names):
        analyze_and_visualize_permutation_importance(best_model, X_test, y_test, feature_names, model_name)

if __name__ == "__main__":
    main()
```

```python
# Modified first code block to work for classification

# Import necessary libraries
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix, plot_confusion_matrix

# Function to detect file encoding
def detect_file_encoding(file_path):
    with open(file_path, 'rb') as file:
        result = chardet.detect(file.read())
    return result['encoding']

# Function to remove gender related columns from dataframe
def remove_gender(df):
    columns_to_drop = [column for column in df.columns if 'M' in column or 'F' in column]
    return df.drop(columns_to_drop, axis=1)


# Function to remove correlated columns from dataframe
# Define a function named 'remove_correlated_columns' that takes a pandas DataFrame 'df' and a threshold value as inputs
def remove_correlated_columns(df, threshold=0.1):
    
    # Compute the correlation matrix of the input DataFrame 'df' using the corr() method, take the absolute value to get the magnitude of the correlation
    corr_matrix = df.corr().abs()
    
    # Generate a list of tuples containing pairs of columns with a correlation coefficient greater than the specified threshold
    high_corr_pairs = [(col1, col2) for col1 in corr_matrix.columns for col2 in corr_matrix.columns if corr_matrix.loc[col1, col2] > threshold and col1 != col2]
    
    # Extract the second column from each pair of 'high_corr_pairs', convert them to a set to remove duplicates, then convert it back to a list
    columns_to_drop = list(set([pair[1] for pair in high_corr_pairs]))
    
    # Drop the highly correlated columns from the original DataFrame 'df' by calling the drop() method and passing in the 'columns_to_drop' list and 'axis=1' to indicate that the columns are to be dropped
    return df.drop(columns_to_drop, axis=1)

#  Function to prepare dataframe with relevant columns
def prepare_data(df):
    
    # Create a new column in the DataFrame 'df' named 'Electoral Division', which is a concatenation of the last three characters of the 'GEOGID' column and the 'GEOGDESC' column
    df['Electoral Division'] = df['GEOGID'].str[-3:] + ' ' + df['GEOGDESC']
    
    # Rename the column 'T10_3_ENGT' to 'Highest Education Level' and the column 'T11_1_TW' to 'Total Hours Worked'
    df = df.rename(columns={'T10_3_ENGT': 'Highest Education Level', 'T11_1_TW': 'Total Hours Worked'})
    
    # Return a new DataFrame with only the columns 'Electoral Division', 'Highest Education Level', and 'Total Hours Worked'
    return df[['Electoral Division', 'Highest Education Level', 'Total Hours Worked']]

# Function to merge multiple dataframes on a common column
def merge_data(dfs, on='Electoral Division'):
    
    # Initialize a new DataFrame 'merged_df' as the first DataFrame in the list 'dfs'
    merged_df = dfs[0]
    
    # Iterate through the remaining DataFrames in the list 'dfs' and merge them with 'merged_df' on the specified column 'on'
    for df in dfs[1:]:
        merged_df = merged_df.merge(df, on=on)
    
    # Return the merged DataFrame
    return merged_df

# Function to keep the first three characters of a string
def keep_first_three_chars(column):
    
    return column[:3]

# Function to convert dataframe columns to int
def convert_to_int(df):
    
    # Iterate through each column in the DataFrame 'df'
    for column in df.columns:
        
        # Check if the column is numeric and not a float
        if pd.api.types.is_numeric_dtype(df[column]) and not pd.api.types.is_float_dtype(df[column]):
            
            # Convert the column to numeric data type, set missing values to the median of the column, and convert it to integer data type
            df[column] = pd.to_numeric(df[column], errors='coerce')
            median = df[column].median()
            df[column] = df[column].fillna(median)
            df[column] = df[column].astype(int)
        
        # If the column is not numeric, replace missing values represented as ".." with NaN and set missing values to the median of the column
        else:
            df[column] = df[column].replace("..", np.nan)
            median = df[column].median()
            df[column] = df[column].fillna(median)
    
    # Return the modified DataFrame
    return df


# Function to process the main CSV file
def process_main_csv():
    
    # Read the CSV file 'main.csv' into a pandas DataFrame 'df', specifying that the thousands separator is a comma
    df = pd.read_csv("main.csv", thousands=',')
    
    # Apply the 'keep_first_three_chars' function to the 'Electoral Division' column of the DataFrame 'df'
    df['Electoral Division'] = df['Electoral Division'].apply(keep_first_three_chars)
    
    # Modify the DataFrame 'df' by converting numeric columns to integer data type and handling missing values
    modified_df = convert_to_int(df)
    
    # Write the modified DataFrame 'modified_df' to a new CSV file named 'main_modified.csv', excluding the index column
    modified_df.to_csv('main_modified.csv', index=False)

def map_income_to_class(income):
    if income <= 20000:
        return 0
    elif 20000 < income <= 40000:
        return 1
    elif 40000 < income <= 60000:
        return 2
    else:
        return 3

# Function to scale the features using StandardScaler
def scale_features(X_train, X_test):
    
    # Initialize a StandardScaler object 'scaler'
    scaler = StandardScaler()
    
    # Fit and transform the training data 'X_train' using the StandardScaler object 'scaler', and store the result in 'X_train_scaled'
    X_train_scaled = scaler.fit_transform(X_train)
    
    # Transform the testing data 'X_test' using the StandardScaler object 'scaler', and store the result in 'X_test_scaled'
    X_test_scaled = scaler.transform(X_test)
    
    # Return the scaled training and testing data as a tuple
    return X_train_scaled, X_test_scaled
    
# Function to split the data into train and test sets
def split_data(df, target_col, test_size=0.2, random_state=42):
    
    # Create a new DataFrame 'X' by dropping the target column from the input DataFrame 'df'
    X = df.drop(target_col, axis=1)
    
    # Create a new Series 'y' by selecting the target column from the input DataFrame 'df'
    y = df[target_col]
    
    # Split the data into training and testing sets using the train_test_split function, and store the results in four variables: X_train, X_test, y_train, y_test
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    
    # Return the training and testing sets as a tuple
    return X_train, X_test, y_train, y_test

# Function to train and evaluate model
def train_and_evaluate_model(model, X_train, X_test, y_train, y_test, model_name):
    # Train the machine learning model using X_train and y_train data
    model.fit(X_train, y_train)
    
    # Predict the target variable for X_test data using the trained model
    y_pred = model.predict(X_test)

    # Calculate accuracy, precision, recall, and F1 score for the predictions
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

    # Return the evaluation metrics, predicted target variable, and trained model
    return accuracy, precision, recall, f1, y_pred, model

# Function that plots relevant metrics on a bar chart
def plot_metrics_bar_chart(metrics, model_names, metric_names):
    # Set the figure size and bar width
    fig, ax = plt.subplots(figsize=(12, 6))
    bar_width = 0.15

    # Create an array of x-axis values for each model
    x = np.arange(len(model_names))

    # Plot a bar for each metric for each model
    for i, metric in enumerate(metric_names):
        ax.bar(x + i * bar_width, [m[i] for m in metrics], width=bar_width, label=metric)

    # Set x-tick positions and labels
    ax.set_xticks(x + bar_width * (len(metric_names) - 1) / 2)
    ax.set_xticklabels(model_names)

    # Set axis labels, limits, and title
    ax.set_xlabel("Models")
    ax.set_ylabel("Score")
    ax.set_ylim([0, 1])
    ax.set_title("Model Performance Metrics")

    # Add a legend
    ax.legend(loc='best')

    # Adjust spacing and display the plot
    plt.tight_layout()
    plt.show()
    
# Function for visualising confusion matrices
def plot_confusion_matrices(best_models, model_names, X_test, y_test):
    # Loop through the best trained models and their corresponding names
    for model, model_name in zip(best_models, model_names):
        # Create a new plot for each model
        fig, ax = plt.subplots(figsize=(6, 6))

        # Generate the confusion matrix for the model and plot it
        plot_confusion_matrix(model, X_test, y_test, ax=ax, cmap='viridis')

        # Set the title for the plot
        ax.set_title(f"Confusion Matrix: {model_name}")

        # Adjust spacing and display the plot
        plt.tight_layout()
        plt.show()

def main():
# Read SAPS2016_ED3409.csv file and remove gender column
    df = pd.read_csv("SAPS2016_ED3409.csv", encoding='ISO-8859-1')
    df = remove_gender(df)
    df.to_csv("SAPS2016_ED3409_no_gender.csv", index=False)

    # Read the modified CSV file and remove correlated columns
    df = pd.read_csv("SAPS2016_ED3409_no_gender.csv")
    df = remove_correlated_columns(df)
    df.to_csv("SAPS2016_ED3409_reduced_columns.csv", index=False)

    # Read the reduced columns CSV file and prepare data
    df = pd.read_csv("SAPS2016_ED3409_reduced_columns.csv")
    df = prepare_data(df)
    df.to_csv('SAPS2016_ED3409_ready_to_merge.csv', index=False)

    # Merge different datasets
    ed_proportion_sector = pd.read_csv("EDProportionSector.csv", encoding='ISO-8859-1')
    ed_social_welfare = pd.read_csv("EDSocialWelfare.csv", encoding='ISO-8859-1')
    ed_pension = pd.read_csv("EDPension.csv", encoding='ISO-8859-1')
    ed_median_gross_income_sd_commuter = pd.read_csv("EDMedianGrossIncomeSDCommuter.csv", encoding='ISO-8859-1')
    ed_median_gross_income_ld_commuter = pd.read_csv("EDMedianGrossIncomeLDCommuter.csv", encoding='ISO-8859-1')
    ed_gross_income_2016 = pd.read_csv("EDGrossIncome2016.csv", encoding='ISO-8859-1')

    # Store the different dataframes in a list
    dfs = [ed_proportion_sector, ed_social_welfare, ed_pension, ed_median_gross_income_sd_commuter, ed_median_gross_income_ld_commuter, ed_gross_income_2016]
    # Merge all the dataframes
    merged_df = merge_data(dfs)
    merged_df.to_csv("MergedData.csv", index=False)

    # Read ready-to-merge dataset and merge with merged dataframe
    df1 = pd.read_csv("SAPS2016_ED3409_ready_to_merge.csv")
    df2 = pd.read_csv("MergedData.csv")

    # Extract only the electoral division from the electoral division column
    df2['Electoral Division Stripped'] = df2['Electoral Division'].apply(lambda x: x.split(',')[0].strip())
    # Merge two dataframes based on electoral division
    merged_df = pd.merge(df1, df2, left_on='Electoral Division', right_on='Electoral Division Stripped', suffixes=('_x', '_y'))

    # Drop unnecessary columns and rename columns
    merged_df.drop(['Electoral Division_x', 'Electoral Division Stripped'], axis=1, inplace=True)
    merged_df.rename(columns={'Electoral Division_y': 'Electoral Division'}, inplace=True)

    # Store the merged dataframe in a csv file
    merged_df.to_csv("main.csv", index=False)

    # Process the main csv file
    process_main_csv()
    
    # Load the modified dataset
    df = pd.read_csv('main_modified.csv')

    # Set the target column as 'Household Income'
    target_col = 'Household Income'

    # Convert the target column to class labels
    df[target_col] = df[target_col].apply(map_income_to_class)
    
    # Remove outliers using IQR method
    df = remove_outliers_iqr(df, target_col)

    # Split the dataset into 3 subsets, scale features, and store subsets in a list
    subsets = []
    for _ in range(3):
        X_train, X_test, y_train, y_test = split_data(df, target_col)
        # Store the column names before scaling
        feature_names = X_train.columns
        X_train, X_test = scale_features(X_train, X_test)
        subsets.append((X_train, X_test, y_train, y_test))
        
    # Define the list of classification models to use
    models = [
        LogisticRegression(random_state=42),
        RandomForestClassifier(random_state=42),
        KNeighborsClassifier(),
        SVC(random_state=42)
    ]

    # Define the corresponding names for each model
    model_names = [
        'Logistic Regression',
        'Random Forest',
        'K-Nearest Neighbors',
        'Support Vector Machines'
    ]

    # Create empty lists for storing the performance metrics and best models
    metrics = []
    best_models = []

    # Loop through each model and train/evaluate it using the training and testing data
    for model, model_name in zip(models, model_names):
        accuracy, precision, recall, f1, y_pred, best_model = train_and_evaluate_model(model, X_train, X_test, y_train, y_test, model_name)

        # Store the best model for further analysis
        best_models.append(best_model)

        # Store the evaluation metrics for the current model
        metrics.append((accuracy, precision, recall, f1))

        # Print the evaluation metrics for the current model
        print(f'{model_name} Metrics:')
        print(f'  Accuracy: {accuracy}\n  Precision: {precision}\n  Recall: {recall}\n  F1 Score: {f1}')

    # Plot the bar chart for performance metrics
    metric_names = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
    plot_metrics_bar_chart(metrics, model_names, metric_names)

    # Plot confusion matrices for each model
    plot_confusion_matrices(best_models, model_names, X_test, y_test)

    
if __name__ == "__main__":
    main()
```
